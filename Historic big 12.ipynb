```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

Get all historical big 12 data from https://basketball.realgm.com/

```python
html=requests.get('https://basketball.realgm.com/ncaa/conferences/Big-12-Conference/3/Awards/Historical')
soup=BeautifulSoup(html.text)
```

```python
## get titles of dataframes

titles = []
for i in str(soup.prettify()).split('</a>'):
    
    # get only tables
    if 'thead' in i:
        
        # split body and head
        for j in i.split('</thead>'):
            
            # header
            if 'thead' in j:
                
                headers = []
                # get each header element
                for head in j.split('</th>'):
                    
                    # get title
                    if 'h2' in head:
                        title = head.split('\n')[2].strip()
                        titles.append(title)
                    elif 'th' in head:
                    
                        headers.append(head.split('\n')[2].strip())
            else:
                
                pass
len(titles)
```

```python
page = pd.read_html('https://basketball.realgm.com/ncaa/conferences/Big-12-Conference/3/Awards/Historical')
```

```python
tables = {}
for i in range(len(titles)):
    tables[titles[i]] = page[i]
```
